# Calculating Cost

```python
import math
```

```python
def total_cost_to_company(a,b,c,d,S,H,cp):
    
    #Economic order Quantity
    
    D = b * 365 # Demand anually calculated by multiplying day consumption with year
    Q = math.sqrt((2 * D * S) / H)
    
    #Reordered point
    ROP = (b * a) + d
    
    Total_Cost = (D/Q * S) + (Q/2 * H) + (D * cp) + (H * d)
    
    return {"EOP":Q,"Total_Cost":Total_Cost,"ROP":ROP}
```

```python
### lets run and calculate total cost by assuming values

a = 80
b = 20
c = 560
d = 30
S = 25
H = 35
cp = 5

total_cost_to_company(a,b,c,d,S,H,cp)
```

# Using Genetic Algo

```python

a = 10  # Lead time in days
b = 5   # Daily consumption
S = 50  # Ordering cost per order
H = 2   # Holding cost per unit per year
C = 10  # Cost per unit



def calculate_total_cost(individual):
    Q, d = individual
    D = b * 365
    EOQ = Q
    daily_demand = b
    ROP = (daily_demand * a) + d
    
    TC = (D / EOQ * S) + (EOQ / 2 * H) + (D * C) + (d * H)
    return TC,
```

```python
import random
import math

# Distributed Evolutionary Algorithms
from deap import base, creator, tools
```

```python
# DEAP setup

# it will create new fitness class called "FitnessMin" and -ve sign indicate it for minimise
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))


creator.create("Individual", list, fitness=creator.FitnessMin)


#Toolbox used to register functions and operators for genetic algorithm.
toolbox = base.Toolbox()

# Chromosome initialization
toolbox.register("EOQ", random.randint, 1, 1000)  # Example range for EOQ
toolbox.register("SafetyStock", random.randint, 1, 100)  # Example range for safety stock

#tuple of functions that will be used to initialize the individual's attributes.
#initializes an individual by cycling through the given functions.

toolbox.register("individual", tools.initCycle, creator.Individual,
                 (toolbox.EOQ, toolbox.SafetyStock), n=1)

# initializes population by repeating individuals.

toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# Genetic operators

toolbox.register("evaluate", calculate_total_cost)
toolbox.register("mate", tools.cxBlend, alpha=0.5)  # blend crossover, alpha==extent of the blending.
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.2)  # gaussian mutation,mean=0,std=1,
#probability for each attribute to be mutated
toolbox.register("select", tools.selTournament, tournsize=3)# tournament selection 3 compete in each tournament
```

```python
population = toolbox.population(n=50)  # list of list consists of (EOQ AND SAFETY STOCK)
population[0:5]   # to visualize the data
```

# Everything in a function

```python
def main():
    population = toolbox.population(n=50)
    
    #crossover probablity,mutation probablity,number of generation
    # 40 times interation happens and evolves population through selection, crossover, and mutation
    crossover_proba, mutation_proba, total_generation = 0.8, 0.4, 60

    # Evaluate the entire population
    fitnesses = map(toolbox.evaluate, population)
    for eoq_safety, fit in zip(population, fitnesses):
        eoq_safety.fitness.values = fit
    
    
    ####### 40 times iteration happens now cosisting of selection, crossover, and mutation
    
    best_fitness_values = []
    safety_stock_values = []
    EOQ_values = []
    
    for g in range(total_generation):
        offspring = toolbox.select(population, len(population))  # selecting based on tournament
        offspring = list(map(toolbox.clone, offspring)) # to ensure original population won't altered
 
        for child1, child2 in zip(offspring[::2], offspring[1::2]):
            if random.random() < crossover_proba:
                toolbox.mate(child1, child2)  # blend cross over
                
                # Deletes fitness values children as altered by crossover and needs to be re-evaluated.
                del child1.fitness.values
                del child2.fitness.values

        for mutant in offspring:
            if random.random() < mutation_proba:  # gaussian mutation used
                toolbox.mutate(mutant)
                
                #Deletes fitness values children as altered by mutation and needs to be re-evaluated
                del mutant.fitness.values

        invalid_eoq_safety = [eoq_safety for eoq_safety in offspring if not eoq_safety.fitness.valid]
        fitnesses = map(toolbox.evaluate, invalid_eoq_safety)
        for eoq_safety, fit in zip(invalid_eoq_safety, fitnesses):
            eoq_safety.fitness.values = fit

        population[:] = offspring  # Replacing current population new offspring for the next generation.
    ############################################################################################
        # for plotting purpose
        fits = [ind.fitness.values[0] for ind in population]
        best_fitness = min(fits)
        best_fitness_values.append(best_fitness)
        
         # safety stock values
        safety_stocks = [ind[1] for ind in population]
        safety_stock_values.append(safety_stocks)
        
         # EOQ values
        EOQ = [ind[0] for ind in population]
        EOQ_values.append(EOQ)
    
    
    best_eoq_safety = tools.selBest(population, 1)[0]
    
    
    print("Best individual is %s, %s" % (best_eoq_safety, best_eoq_safety.fitness.values))
    
    plt.figure(figsize=(10, 5))

    # Plot total cost vs safety stock for the final population
    final_population = [ind for ind in population]
    final_costs = [ind.fitness.values[0] for ind in final_population]
    final_EOQs = [ind[0] for ind in final_population]
    final_safety_stocks = [ind[1] for ind in final_population]

    plt.subplot(1, 2, 1)
    plt.scatter(final_EOQs, final_costs)
    plt.xlabel('EOQ')
    plt.ylabel('Total Cost')
    plt.title('Total Cost vs EOQ')
    
    plt.subplot(1, 2, 2)
    plt.plot(best_fitness_values)
    plt.xlabel('Generation')
    plt.ylabel('Best Fitness Value')
    plt.title('Evolution of Best Fitness Value')
    plt.show()
```

```python
main()
```
