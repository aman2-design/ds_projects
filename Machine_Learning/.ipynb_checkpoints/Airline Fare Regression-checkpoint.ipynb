# 5. Feature Engineering

```python
df = df.drop("flight",axis = 1)
df["duration"] = df["duration"].astype('int')
df["days_left"] = df["days_left"].astype('int')
df
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoded_data = pd.get_dummies(df, columns = ['airline', 'source_city','departure_time','stops','arrival_time',
                                                     'destination_city','class'])
```

```python
one_hot_encoded_data
```

```python
one_hot_encoded_data.columns
```

### Outliers Analysis

```python
# outlier analysis will be on below 3 continuous features

df["duration"] = df["duration"].astype('int')
df["days_left"] = df["days_left"].astype('int')
df["price"] = df["duration"].astype('int')
```

```python
fig,ax = plt.subplots(nrows=1 , ncols=2 , figsize = (20,5))
sns.boxplot(x=df.price , ax = ax[0])
sns.boxplot(x=df.days_left , ax = ax[1])
```

```python
fig,ax = plt.subplots(nrows=2 , ncols=2 , figsize = (20,5))
sns.scatterplot(x = df["price"] , y = np.zeros_like(df.price) , ax = ax[0][0])
sns.scatterplot(x = df["days_left"] , y = np.zeros_like(df.days_left) , ax = ax[0][1])
sns.scatterplot(x = df["duration"] , y = np.zeros_like(df.days_left) , ax = ax[1][1])
```

```python
#No Such Outliers to remove
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
one_hot_encoded_data[['duration', 'days_left']] = scaling.fit_transform(one_hot_encoded_data[['duration', 'days_left']])
```

```python
one_hot_encoded_data.head(4)
```

```python
one_hot_encoded_data.describe()
```

```python
df_processed = one_hot_encoded_data
```

### 6.2 Multicollinearity

```python
x = df_processed.drop("price",axis = 1)
y = df_processed["price"]
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor
estimator = RandomForestRegressor()
selector = RFE(estimator, n_features_to_select=10, step=1)
selector.fit(x,y)
```

```python
X_train_selected = selector.transform(x)
```

```python
X_train_selected
```

```python
mask = selector.get_support()
plt.matshow(mask.reshape(1,-1),cmap='gray_r')
plt.xlabel("Sample index")
plt.yticks(())
plt.show()
```
