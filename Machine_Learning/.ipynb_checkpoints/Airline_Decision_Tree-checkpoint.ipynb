```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
airline_data = pd.read_csv("Invistico_Airline.csv")
```

```python
airline_data.head(3)
```

```python
airline_data.shape
```

```python
airline_data.isnull().sum()
```

```python
# filling null values

airline_data["Arrival Delay in Minutes"] = airline_data["Arrival Delay in Minutes"
                                                       ].fillna(airline_data["Arrival Delay in Minutes"].mean())
```

```python
airline_data.isnull().sum()
```

```python
airline_data.info()
```

```python
# converting categorical variable to continuous

airline_data["satisfaction"] = airline_data["satisfaction"].apply(lambda x:1 if x == "satisfied" else 0)
airline_data["Gender"] = airline_data["Gender"].apply(lambda x:1 if x == "Male" else 0)
airline_data["Customer Type"] = airline_data["Customer Type"].apply(lambda x:1 if x == "Loyal Customer" else 0)
airline_data["Type of Travel"] = airline_data["Type of Travel"].apply(lambda x:1 if x == "Business travel" else 0)
airline_data["Class"] = airline_data["Class"].map({"Business":2,"Eco Plus":1,"Eco":0})
```

```python
x = airline_data_1.drop("satisfaction",axis = 1)
```

```python
y = airline_data_1["satisfaction"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
print("X_train:",X_train.shape)
print("y_train:",y_train.shape)
print("X_test:",X_test.shape)
print("y_test:",y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
clf.feature_importances_
```

```python
X_train.columns
```

```python
#Feature importance

Feature_importance = dict(zip(X_train.columns,clf.feature_importances_))
```

```python
Feature_importance
```

```python
pd.DataFrame(Feature_importance.values(),Feature_importance.keys(),columns=['Importance'])
```

```python
type(Feature_importance.values())
values = list(Feature_importance.values())
keys = list(Feature_importance.keys())
```

```python
plt.figure(figsize = (20,5))
ax = sns.barplot(x = keys,y = values)
plt.title("Feature Importance Graph")
plt.xlabel("Features")
plt.xlabel("Importance")
plt.bar_label(ax.containers[0])
```

```python
clf.predict_proba(X_test)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
from sklearn import metrics
```

```python
print("accuracy:",metrics.accuracy_score(y_test,y_pred))
print("precision:",metrics.precision_score(y_test,y_pred))
print("recall:",metrics.recall_score(y_test,y_pred))
print("f1_score:",metrics.f1_score(y_test,y_pred))
```

```python
confusion_matrix = metrics.confusion_matrix(y_test,y_pred)
confusion_matrix
```

```python
disp = metrics.ConfusionMatrixDisplay(confusion_matrix)
disp.plot()
```
