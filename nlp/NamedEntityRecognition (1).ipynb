# Train, Validation, Test split

```python
train_data = 0.7
validation_data = 0.10
test_data = 0.20
```

```python

# Took reference form chat GPT reference


import numpy as np

def split_data(X, y, train_ratio=0.7, val_ratio=0.1, test_ratio=0.2):

    assert train_ratio + val_ratio + test_ratio == 1, "Ratios should sum up to 1"
    assert len(X) == len(y), "Input features and labels must have the same length"

    # Shuffle the data
    indices = np.random.permutation(len(X))
    X_shuffled = X[indices]
    y_shuffled = y[indices]

    # Calculate split indices
    total_len = len(X)
    train_end = int(total_len * train_ratio)
    val_end = train_end + int(total_len * val_ratio)

    # Split the data
    train_X, train_y = X_shuffled[:train_end], y_shuffled[:train_end]
    val_X, val_y = X_shuffled[train_end:val_end], y_shuffled[train_end:val_end]
    test_X, test_y = X_shuffled[val_end:], y_shuffled[val_end:]

    return train_X, train_y, val_X, val_y, test_X, test_y


train_X, train_y, val_X, val_y, test_X, test_y = split_data(X_preprocessed, Y_preprocessed)

print("Train data shapes - X:", train_X.shape, "y:", train_y.shape)
print("Validation data shapes - X:", val_X.shape, "y:", val_y.shape)
print("Test data shapes - X:", test_X.shape, "y:", test_y.shape)
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((train_X, train_y))
val_dataset = tf.data.Dataset.from_tensor_slices((val_X, val_y))
test_dataset = tf.data.Dataset.from_tensor_slices((test_X, test_y))
```

```python
train_X[0]
```

```python

#A batch size of 132 means that 132 samples will be processed together in each iteration of training

BATCH_SIZE = 132
SHUFFLE_BUFFER_SIZE = 132

train_dataset = train_dataset.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
val_dataset = val_dataset.batch(BATCH_SIZE)
test_dataset = test_dataset.batch(BATCH_SIZE)
```

# Model Building

```python
# Creating and neural network


#In natural language processing (NLP) tasks, words are often represented as dense vectors in an embedding space, 
#and embedding_dim specifies the size of these vectors.

embedding_dim = 300
maxlen = 110 # For each sentence length is 110(less than that will be padded and more will be truncated)
max_words = 36000 # Total words in vocabulaory limit
num_tags = len(tags)  # total distinct tags

model = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(max_words, embedding_dim, input_length=maxlen),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(units=100, activation='tanh', return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(units=100, activation='tanh', return_sequences=True)),
    tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(num_tags, activation='softmax'))
])


#A time distributed dense layer takes a batch size by sequence length by input size array and produces 
#a batch size by sequence length by number of classes size array

#model consists of an embedding layer followed by two bidirectional LSTM layers and a time-distributed 
#dense layer for multi-label classification, where each output corresponds to a tag or label.
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])


#This is the loss function used for multi-class classification tasks where the target labels are integers. 
#It calculates the cross-entropy loss between the true labels and the predicted probabilities. 
```

```python
history = model.fit(train_dataset,
                    validation_data=val_dataset,
                    epochs=30)
```

```python
model.evaluate(test_dataset)
```
