```python
from urllib.request import Request,urlopen
from bs4 import BeautifulSoup
import html5lib
import requests
def get_article_text(df):
    
    
    df["Article_Text_no_processed"] = ""
    article_text_list = []
    for i in df.index:
        link = df["Link"][i]
        req = Request(link,headers={'User-Agent':'Mozilla/5.0'})
        article = urlopen(req).read()

        parsed_article = BeautifulSoup(article,'html5lib')

        paragraphs = parsed_article.find_all('p')
        #print(paragraphs)

        article_text = ""

        for p in paragraphs:
            article_text += p.text
            df["Article_Text_no_processed"][i] = article_text
            
    return df       
    
        
        
```

```python
import re

df_priorities["Article_Text_processed"] = ""
def preprocessing(df):
    for i in df.index:
        df_priorities["Article_Text_processed"][i] = re.sub(r'\[[0-9]*\]', ' ', df_priorities["Article_Text_no_processed"][i])
        df_priorities["Article_Text_processed"][i] = re.sub(r'\s+', ' ', df_priorities["Article_Text_processed"][i])
        # Removing special characters and digits
        df_priorities["Article_Text_processed"][i] = re.sub('[^a-zA-Z]', ' ', df_priorities["Article_Text_processed"][i])
        df_priorities["Article_Text_processed"][i] = re.sub(r'\s+', ' ', df_priorities["Article_Text_processed"][i])
    return df    
```

```python
df_priorities_processed = preprocessing(df_priorities)
df_priorities_processed
```

### One way of summarization

```python

import nltk

def weighted_count(df):
    df["Weighted_count"] = {}
    stopwords = nltk.corpus.stopwords.words('english')
    
    # to get word frequency
    for i in df_priorities_processed.index:
        word_frequencies = {}
        for word in nltk.word_tokenize(df_priorities_processed["Article_Text_processed"][i]):
            if word not in stopwords and len(word)>1:
                if word not in word_frequencies.keys():
                    word_frequencies[word] = 1
                else:
                    word_frequencies[word] += 1
        maximum_frequncy = max(word_frequencies.values())
        
        # to get weighted count
        for word in word_frequencies.keys():
            word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)
        df["Weighted_count"][i] = word_frequencies
    return df   
            
    
```

```python
def sentence_scores(df):
    df["sentence_scores"] = {}
    for i in df_priorities_processed.index:
        
        sentence_list = nltk.sent_tokenize(df_priorities_processed["Article_Text_no_processed"][i])

        sentence_scores = {}
        for sent in sentence_list:
            for word in nltk.word_tokenize(sent.lower()):
                if word in df_priorities_processed["Weighted_count"][i].keys():
                    if len(sent.split(' ')) < 30:
                        if sent not in sentence_scores.keys():
                            sentence_scores[sent] = df_priorities_processed["Weighted_count"][i][word]
                        else:
                            sentence_scores[sent] += df_priorities_processed["Weighted_count"][i][word]
        df["sentence_scores"][i] = sentence_scores                  
    return df                       
    
```
